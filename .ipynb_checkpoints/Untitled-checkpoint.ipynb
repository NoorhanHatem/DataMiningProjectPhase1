```python
import numpy as np
import pandas as pd
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
house_data = pd.read_csv("house_price.csv")
house_data.head()
```

```python
rooms = list(house_data['rooms'])
area = list(house_data['area'])
price = list(house_data['price'])

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.scatter3D(area, rooms, price, color = "blue")
plt.title("House Price 3D Scatter Plot")

ax.set_xlabel('Area')
ax.set_ylabel('Rooms')
ax.set_zlabel('')

plt.show()
```

```python
X = house_data[['area', 'rooms']].values.reshape(-1,2)
Y = house_data['price']

df2 = pd.DataFrame(X, columns=['area', 'rooms'])
df2['price'] = pd.Series(Y)

model = smf.ols(formula='price ~ area + rooms', data=df2)
results_formula = model.fit()
results_formula.params

x_surf, y_surf = np.meshgrid(np.linspace(df2.area.min(), df2.area.max(), 100),np.linspace(df2.rooms.min(), df2.rooms.max(), 100))
onlyX = pd.DataFrame({'area': x_surf.ravel(), 'rooms': y_surf.ravel()})
fittedY=results_formula.predict(exog=onlyX)

fittedY=np.array(fittedY)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df2['area'],df2['rooms'],df2['price'],c='red', marker='o', alpha=0.5)
ax.plot_surface(x_surf,y_surf,fittedY.reshape(x_surf.shape), color='b', alpha=0.3)
ax.set_xlabel('Area')
ax.set_ylabel('Rooms')
ax.set_zlabel('Price')
plt.show()
```

```python
results_formula.outlier_test(method='simes-hochberg', alpha=0.05, labels=None, order=False, cutoff=0.5)
```

```python
import pandas
from sklearn import linear_model

X = house_data[['area', 'rooms']]
Y = house_data['price']

regr = linear_model.LinearRegression()
regr.fit(X.values, Y.values)

print("Area: 2567    ", "Rooms:5    ", "Predicted Price:", regr.predict([[2567, 5]]))
print("Area: 1200    ", "Rooms:2    ", "Predicted Price:", regr.predict([[1200, 2]]))
print("Area: 852     ", "Rooms:5    ", "Predicted Price:", regr.predict([[852, 5]]))
print("Area: 1852    ", "Rooms:2    ", "Predicted Price:", regr.predict([[1852, 2]]))
print("Area: 1203    ", "Rooms:3    ", "Predicted Price:", regr.predict([[1203, 3]]))
```
